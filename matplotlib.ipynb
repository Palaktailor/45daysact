```python
x=[1,2,3,4,5]
y=[1,10,9,30,25]
plt.plot(x,y,marker='o',ms=10,
         mfc="yellow",mec="green",color="black",lw=3,ls="--",alpha=0.3,label="lineplot") 
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show
```

```python
# line plot and scatter plot -- numerical data
# bar graph -- categorical(alphabetic)
x=["a","b","c","d"]
y = [10,20,30,40]
plt.bar(x,y)
plt.show
```

```python
x=["a","b","c","d"]
y = [10,20,30,40]

colours = ["red","green","blue","yellow"]
hatch_patterna = ["//","\\\\","xx","oo"]
plt.bar(x,y,color =colours , hatch = hatch_patterna , label = x,width=0.3,
        edgecolor ="black",alpha = 0.7,
        linewidth = 3)# by width uyou can control you bars width --alpha lighten or darken the colour

plt.legend() 
plt.show
```

```python
plt.barh(x,y)
plt.legend()
plt.show
```

```python
#histogram --distribution of data in intervals(bins)
arr = np.random.randint(20000,40000,100)
plt.hist(arr)
plt.show
```

```python
##piechart -- percentage
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
plt.pie(y,labels = x)
plt.show
```

```python
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
plt.pie(y,labels = x,autopct="%1.2f%%",shadow = True) # autopct -- to show percentage
plt.show
```

```python
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
explodes = (0.1,0.1,0.1,0.1) # to dur dur krne ke liye besically kaam aata hai 
plt.pie(y,labels = x,autopct="%1.2f%%",shadow = True,explode = explodes) # autopct -- to show percentage
plt.legend()
plt.show
```

```python
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
explodes = (0.1,0.1,0.1,0.1) # to dur dur krne ke liye besically kaam aata hai 
plt.pie(y,labels = x,autopct="%1.2f%%",shadow = True,explode = explodes) # autopct -- to show percentage
plt.legend()
plt.savefig("piechart.png")
plt.show
```

```python
##subplot
import warnings
warnings.filterwarnings("ignore") # to iggnore the warnings
x = np.linspace(0,10,100)
plt.figure(figsize=(10,5))# figsize is used to increase the size of graph
plt.subplot(2,2,1)#2rows .2 columns,1st graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2)#2rows .2 columns,2nd graph
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3)#2rows .2 columns,3rd graph
plt.plot(x,np.tan(x))
plt.title("tan")
plt.subplot(2,2,4)#2rows .2 columns,4th graph
plt.plot(x,np.log(x))
plt.title("log")

plt.show()
```
