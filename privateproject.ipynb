**MIssing Value Treatment**-- since we have so many missing values

```python
def missing_value_treatment(s):
    if s.dtype == "object":
        s = s.fillna(s.mode())
    else:
        s = s.fillna(s.median())
    return s
        
```

```python
df = df.apply(missing_value_treatment)
```
