```python
# outliers handle
#q1 = 25 percentile --- Q2 = 50 --------Q3 = 70
q1 = df["total_bill"].quantile(0.25)
q3 = df["total_bill"].quantile(0.75)
iqr = q3-q1
lower_limit = q1-1.5*iqr
upper_limit = q3+1.5*iqr
print(lower_limit ,upper_limit)
df[df["total_bill"]<lower_limit]
```

```python
##outliers handle
q1 = df['total_bill'].quantile(0.25)
q3 = df['total_bill'].quantile(0.75)
iqr=q3-q1  #inter quartile range
lower_limit=q1-1.5*iqr #lower limit
upper_limit=q3+1.5*iqr #upper limit
print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
#df[df["tip"]<lower_limit]  #outliers below lower limit
df[df["total_bill"]>upper_limit]  #outliers above upper limit
#removing outliers
df=df[(df["total_bill"]>lower_limit) & (df["total_bill"]<upper_limit)]
sns.boxplot(df["total_bill"])
```
